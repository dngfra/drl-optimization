### Noise 0.1

```python
weights, scores = utils.get_results('20181228171611')
```

```python
weights_hidden = np.empty((len(weights), len(weights[0]), weights[0][0][0].shape[0]*weights[0][0][0].shape[1]))
for i, w in enumerate(weights):
    for j, ww in enumerate(w):
        weights_hidden[i][j] = ww[0].reshape(-1)
```
